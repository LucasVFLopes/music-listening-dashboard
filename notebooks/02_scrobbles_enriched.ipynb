```python
import pandas as pd

df = pd.read_csv("scrobbles_dashboard.csv")
df.shape
```

```python
df.columns
```

```python
df["played_at_utc"] = pd.to_datetime(df["played_at_utc"])
```

```python
df["played_at_utc"].head()
```

```python
df["played_at_br"] = df["played_at_utc"].dt.tz_convert("America/Sao_Paulo")
```

```python
df[["played_at_utc", "played_at_br"]].head(5)
```

```python
df["hour"] = df["played_at_br"].dt.hour
df["day_of_week"] = df["played_at_br"].dt.day_name()
df["day_of_week_num"] = df["played_at_br"].dt.weekday
df["is_weekend"] = df["day_of_week_num"] >= 5
```

```python
def period_of_day(hour):
    if 0 <= hour <= 5:
        return "madrugada"
    elif 6 <= hour <= 11:
        return "manha"
    elif 12 <= hour <= 17:
        return "tarde"
    else:
        return "noite"

df["period_of_day"] = df["hour"].apply(period_of_day)
```

```python
df[[
    "played_at_utc",
    "played_at_br",
    "hour",
    "period_of_day",
    "day_of_week",
    "macro_genre",
    "energy_level"
]].head(10)
```

```python
df.to_csv(
    "scrobbles_dashboard_time_enriched.csv",
    index=False,
    encoding="utf-8"
)
```
